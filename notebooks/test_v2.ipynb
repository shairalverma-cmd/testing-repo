```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
import json
import os
import numpy as np
import shutil # For sorting files
```

```python
try:
    model_v2 = tf.keras.models.load_model('../models/model_v2.h5')
    print("Loaded model_v2.h5")
except Exception as e:
    print(f"Failed to load model_v2.h5, \n error: {e}")
```

```python
# --- ML Code: Preprocessing ---
IMG_SIZE = (128, 128)
BATCH_SIZE = 32

print("--- User 1 Data Structure ---") 
!dir ..\\data\\user1
```

```python
DATA_DIR_USER1 = '../data/user1/training_set/training_set'

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
train_generator_u1 = datagen.flow_from_directory(
    DATA_DIR_USER1,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary', # Changed to binary for Cat vs Dog
    subset='training'
)
validation_generator_u1 = datagen.flow_from_directory(
    DATA_DIR_USER1,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary', # Changed to binary
    subset='validation'
)
```

```python
try:

    print("Testing model_v2 on User 1's validation data...")
    results_v2_on_u1_data = model_v2.evaluate(validation_generator_u1)

    test_metrics_v2_u1 = {
        'test_loss': results_v2_on_u1_data[0],
        'test_accuracy': results_v2_on_u1_data[1]
    }

except Exception as e:
    print(f"Error testing model_v2: {e}")
    print("This might be a domain shift problem or a subtle model mismatch.")
```

```python

# Make sure the directory exists
os.makedirs('results', exist_ok=True)

# Save the file
with open('../results/test_v2_user1.json', 'w') as f:
    json.dump(test_metrics_v2_u1, f, indent=4)

print("Saved test_v2_user1.json to results/")
```
